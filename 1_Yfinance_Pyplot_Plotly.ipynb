### Dealing with multiple tickers

```python
data = yf.download("AMZN AAPL GOOG", start="2017-01-01", end="2017-04-30")
data
```

This code retrieves stock information for multiple companies using the yfinance library and organizes the data into a single pandas DataFrame.

These represent Apple (AAPL), Google (GOOG), Amazon (AMZN), Bank of America (BAC), and Boeing (BA).



```python
tickers_list = ["aapl", "goog", "amzn", "BAC", "BA"] # example list
tickers_data= {} # empty dictionary A dictionary tickers_data is created to store the ticker name 
                 # as the key and its corresponding DataFrame as the value.

for ticker in tickers_list:
    ticker_object = yf.Ticker(ticker)

    #convert info() output from dictionary to dataframe
    temp = pd.DataFrame.from_dict(ticker_object.info, orient="index")
    temp.reset_index(inplace=True)
    temp.columns = ["Attribute", "Recent"]
    
    # add (ticker, dataframe) to main dictionary
    tickers_data[ticker] = temp

# At this point, tickers_data is a dictionary where:
#       Each key is a stock ticker.
#       Each value is a DataFrame of attributes for that stock.

tickers_data
```

```python
#  Concatenates all DataFrames stored in tickers_data into a single large DataFrame.
combined_data = pd.concat(tickers_data)
combined_data = combined_data.reset_index()
combined_data
```

```python
# delete the unnecessary “level_1” column and clean up the column names
del combined_data["level_1"] # clean up unnecessary column
combined_data.columns = ["Ticker", "Attribute", "Recent"] # update column names

combined_data
```

```python
# This line selects only the rows where the column "Attribute" has the value "fullTimeEmployees" and
# reset the index

employees = combined_data[combined_data["Attribute"]=="fullTimeEmployees"].reset_index()

# Since .reset_index() keeps the old index as a separate column ("index"), this line deletes it to clean up the DataFrame.

del employees["index"]  

employees
```

```python
employees_sorted = employees.sort_values('Recent',ascending=False)
employees_sorted
```
